```python
import pandas as pd
import matplotlib.pyplot as plt

# load the dataset into a pandas DataFrame
df = pd.read_csv('g2m_cab.csv')

# create separate scatter plots for Yellow Cab and Pink Cab
yellow_cab_df = df[df['Company'] == 'Yellow Cab']
pink_cab_df = df[df['Company'] == 'Pink Cab']

plt.rcParams['figure.figsize'] = [16, 6]  # set figure size

plt.subplot(1, 2, 1)  # create the first subplot for Yellow Cab
plt.scatter(yellow_cab_df['Price Charged'], yellow_cab_df['KM Travelled'], c=yellow_cab_df['Gender'].map({'Male': 'blue', 'Female': 'red'}))
plt.title('Yellow Cab: Price Charged vs. Distance Traveled, Colored by Gender')
plt.xlabel('Price Charged')
plt.ylabel('Distance Traveled (km)')

plt.subplot(1, 2, 2)  # create the second subplot for Pink Cab
plt.scatter(pink_cab_df['Price Charged'], pink_cab_df['KM Travelled'], c=pink_cab_df['Gender'].map({'Male': 'blue', 'Female': 'red'}))
plt.title('Pink Cab: Price Charged vs. Distance Traveled, Colored by Gender')
plt.xlabel('Price Charged')
plt.ylabel('Distance Traveled (km)')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# load the dataset into a pandas DataFrame
df = pd.read_csv('g2m_cab.csv')

# group the data by city and company, and count the number of rides in each group
rides_by_city = df.groupby(['City', 'Company'])['Transaction ID'].count().reset_index()

# create a pivot table with the number of rides for each company in each city
pivot_rides_by_city = rides_by_city.pivot(index='City', columns='Company', values='Transaction ID')

# normalize the data by dividing each value by the total number of rides in the city
pivot_rides_by_city_norm = pivot_rides_by_city.div(pivot_rides_by_city.sum(axis=1), axis=0)

# create a stacked bar chart of the proportion of rides for each company in each city
plt.rcParams['figure.figsize'] = [12, 8]  # set figure size
pivot_rides_by_city_norm.plot(kind='bar', stacked=True)
plt.title('Proportion of Pink Cab and Yellow Cab Rides in Each City')
plt.xlabel('City')
plt.ylabel('Proportion of Rides')
plt.legend(title='Company')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# load the dataset into a pandas DataFrame
df = pd.read_csv('g2m_cab.csv')

# select the numeric columns
numeric_cols = ['KM Travelled', 'Price Charged', 'Cost of Trip', 'Age', 'Income (USD/Month)', 'Population', 'Users']

# compute the correlation matrix
corr_matrix = df[numeric_cols].corr()

# create a heatmap of the correlation matrix
plt.rcParams['figure.figsize'] = [10, 8]  # set figure size
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix of Numeric Variables')
plt.show()
```

Do some of the variables have strong correlation? - Yes, price charged and km traveled have strong correlation - which makes sense

Do customers prefer card payments to cash? - Yes, it's the case for both companies

```python
import pandas as pd
import matplotlib.pyplot as plt

# load the dataset into a pandas DataFrame
df = pd.read_csv('g2m_cab.csv')

# compute the total revenue and cost of trip for each company
totals = df.groupby('Company').agg({'Price Charged': 'sum', 'Cost of Trip': 'sum'}).reset_index()

# create a stacked bar chart of the total revenue and cost of trip for each company
plt.rcParams['figure.figsize'] = [12, 8]  # set figure size
companies = totals['Company'].values
revenue = totals['Price Charged'].values
cost = totals['Cost of Trip'].values
plt.bar(companies, revenue, label='Revenue')
plt.bar(companies, cost, bottom=revenue, label='Cost of Trip')
plt.title('Total Revenue and Cost of Trip by Company')
plt.xlabel('Company')
plt.ylabel('Amount')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
g2m_cab = pd.read_csv('g2m_cab.csv')

# Convert 'Date of Travel' column to datetime format
g2m_cab['Date of Travel'] = pd.to_datetime(g2m_cab['Date of Travel'])

# Create new columns for day of the week and hour of the day
g2m_cab['Day of Week'] = g2m_cab['Date of Travel'].dt.dayofweek
g2m_cab['Hour of Day'] = g2m_cab['Date of Travel'].dt.hour

# Group the rides by hour and day of the week
rides_by_hour_day = g2m_cab.groupby(['Hour of Day', 'Day of Week'])['Transaction ID'].count().reset_index()

# Pivot the data to create a matrix of rides by hour and day of the week
rides_by_hour_day_pivot = rides_by_hour_day.pivot('Hour of Day', 'Day of Week', 'Transaction ID')

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(rides_by_hour_day_pivot, cmap='YlGnBu', annot=True, fmt='g', linewidths=.5)
plt.title('Rides by Hour and Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Hour of Day')
plt.show()
```

Is there more demand for cabs during weekend? - Yes, the highest demand is in Friday, then Saturday and then Sunday
